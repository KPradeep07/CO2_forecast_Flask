```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
df_ar = pd.read_excel("CO2 dataset.xlsx")
df_ar.head()
```

```python
index = pd.date_range(start="1800", end="2015", freq="A-DEC")
index
```

```python
df_ar.drop("Year", axis=1, inplace=True)
```

```python
df_ar.set_index(index, inplace=True)
```

```python
df_ar
```

```python
model_arima_final = ARIMA(df_ar, order = (10, 1, 15)).fit()
```

```python
fitted = model_arima_final.predict(start = 0, end = 214)
```

```python
plt.figure(figsize=(12,10))
plt.plot(df_ar, color='green', label='Train')
plt.plot(fitted, color='magenta', label='Forecast')
plt.title("Foecating by ARIMA")
plt.xlabel("year")
plt.ylabel("CO2 level")
plt.legend()
plt.show()
```

```python
forecast_final = model_arima_final.predict(start=215, end = 265)
forecast_final.head()
```

```python
plt.figure(figsize=(10,8))
plt.plot(df_ar, color='green', label='train')
plt.plot(forecast_final, color='Red', label='forecast')
plt.title("Foecating by ARIMA")
plt.xlabel("year")
plt.ylabel("CO2 level")
plt.legend()
plt.show()
```
